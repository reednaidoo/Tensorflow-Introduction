# Importing the dataset

```r
mnist <- dataset_mnist()
```

```r
x_train <- mnist$train$x
y_train <- mnist$train$y
x_test <- mnist$test$x
y_test <- mnist$test$y
```

```r
y_train
```

```r
# reshape
x_train <- array_reshape(x_train, c(nrow(x_train), 784))
x_test <- array_reshape(x_test, c(nrow(x_test), 784))
# rescale
x_train <- x_train / 255
x_test <- x_test / 255
```

```r
x_train
```

```r
y_train <- to_categorical(y_train, 10)
y_test <- to_categorical(y_test, 10)
```

# Building the model

```r
model <- keras_model_sequential() 
model %>% 
  layer_dense(units = 256, activation = 'relu', input_shape = c(784)) %>% 
  layer_dropout(rate = 0.4) %>% 
  layer_dense(units = 128, activation = 'relu') %>%
  layer_dropout(rate = 0.3) %>%
  layer_dense(units = 10, activation = 'softmax')
```

```r
summary(model)
```

```r
model %>% compile(
  loss = 'categorical_crossentropy',
  optimizer = optimizer_rmsprop(),
  metrics = c('accuracy')
)
```

```r

history <- model %>% fit(
  x_train, y_train, 
  epochs = 20, batch_size = 128, 
  validation_split = 0.2
)
```

```r

mutliclass_macro_auc <- function(preds, actual){
    options(warn=-1)
    # prespecifying the number of classes and observations
    n_classes <- length(levels(as.factor(actual)))
    n_obs <- length(actual)
    
    # a matrix consisting of class association (with 1's)
    preds_dat <- data.frame(to_categorical((preds)))
    actual_dat <- data.frame(to_categorical((actual)))


    # creating an empty list for which the auc object is stored 
    empty <- list()
    
    # iterating through the classes to compute auc statistics of each correct class classification 
        for(i in 1:n_classes){
        empty[i] <- suppressMessages(list(roc(preds_dat[,i], actual_dat[,i])))
        }
    
    # creating each class AUC 
    all_auc <- c()
        for(j in 1:n_classes){
        all_auc[j] <- c(empty[[j]]$auc, all_auc)
        }
    
    # creating each class sensitivity 
    all_sens <- c()
        for(k in 1:n_classes){
        all_sens[k] <- c(empty[[k]]$sensitivities[2], all_sens)
        }
    
    # creating each class specificity 
    all_spec <- c()
        for(l in 1:n_classes){
        all_spec[l] <- c(empty[[l]]$specificities[2], all_spec)
        }
    
    
    # dataframe output  
    output_dataframe <- data.frame(cbind(c(1:n_classes), all_auc, all_sens, all_spec))
    colnames(output_dataframe) <- c("classes", "AUC", "Sensitivities", "Specificities")
    

    
    # Macro AUC plot output 
    mean_sens <- mean(all_sens)
    mean_spec <- mean(all_spec)
    
    sens <- c(0, mean_sens, 1)
    spec <- c(1, mean_spec, 0)
    
    ggdat <- data.frame(cbind(sens, spec))
    
    
    options(repr.plot.width = 15, repr.plot.height = 7)
    plot <- ggplot(data = ggdat) +
    geom_smooth(aes(x = c(spec), y = c(sens), col = "ROC AUC"), formula = "y~x", method = 'loess') + 
    theme_linedraw(base_size = 20) + 
    labs(title = "Area under the curve", col = "Curves")+
    xlab(label = "1 - Specificity")+
    ylab(label = "Sensitivity")+
    scale_colour_manual(values = c(3))
    

    
    list("ROC stats" = output_dataframe,
        "Macro AUC" = mean(c(output_dataframe[,2])), 
        plot)


    
}

# adding 1 because our observations technically start at zero
mutliclass_macro_auc((preds_vec), (ytest_vec))
```
